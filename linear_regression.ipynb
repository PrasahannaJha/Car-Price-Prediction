Approach for handling 22 unique entries of carcompany column - carcompany name seem to be a good driver of price. It is important to transform these variables into manageable number of variables otherwise 
1. we might have to add 21 dummy variables to capture it. Adding 21 dummy variables having only zero and one value is not good for the model as the overall number of observations is also very small. Lets group them in highpricedbrand, midpricedbrand and lowpricedbrand groups.
2. brand wise price variation is a known fact. so, it will be business interpretable as well when say porsche is in the highpriced brand group (irrespective of other features). It essentially means addition or deletion of car features will not change the fact that porsche is costlier. But, Porsche's brand might make some features appear as poor driver of price.

Following is the obvious grouping into pricebrands as per the boxplot:
> highpricedbrand = ['bmw', 'jaguar', 'buick' and 'porsche'] <br>
> midpricedbrand = ['alpha-romero', 'audi', 'mercury', 'peugeot', 'saab', 'volvo'] <br>
> lowpricedbrand = ['chevrolet', 'dodge', 'honda', 'isuzu', 'mazda', 'mitsubishi', 'nissan', 'plymouth', 'renault', 'subaru', 'toyota', 'volkswagen'] 

```python
# replacing the car company name by the pricebrand categories 
car_price.carcompany = car_price.carcompany.replace(['bmw', 'jaguar', 'buick', 'porsche'], 'highpricedbrand')
car_price.carcompany = car_price.carcompany.replace(['alfa-romero', 'audi', 'mercury', 'peugeot', 'saab', 'volvo'], 'midpricedbrand')
car_price.carcompany = car_price.carcompany.replace(['chevrolet', 'dodge', 'honda', 'isuzu', 'mazda', 'mitsubishi', 'nissan', 'plymouth', 'renault', 'subaru', 'toyota', 'volkswagen'], 'lowpricedbrand')
```

#### Step4 - Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# initiate the randon state so that train and test split is always the same

df_train, df_test = train_test_split(car_price, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
# rescaling the features with minmax scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables

num_vars =['symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg', 'price']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train.head()
```

```python
### creating X and y sets for model building

y_train = df_train.pop('price')
X_train = df_train
```

#### Step5 - Building linear model

Approach - Since the number of columns is very high we will first use RFE to filter top 20 columns and then clean it up manually

```python
# Importing necessary packages for RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 20
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 20)             
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# top 20 columns that are relevant as per RFE
col = X_train.columns[rfe.support_]
col
```

```python
# rest of the columns as per RFE
col1 = X_train.columns[~rfe.support_]
col1
```

leveraging statsmodel to look at the statistical values

```python
# Creating X_train_rfe dataframe leveraging the variable selected in RFE. 
# also added 'rwd' in the train set as 'rwd' is a driver based on domain knowledge
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# building linear model
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

**Model summary**
1. R2 of the model is good
2. Adjusted R2 is also good
3. prob (F stats) ~ 0: significant prove that the model is not by chance
4. wheelbase is insiginificant in presence of other variables. 

```python
# Checking VIF
# checking the columns in X-train_rfe

X_train_rfe.columns
```

```python
# dropping the constant added for model building and assigning it to a new dataframe - X_train_rfe1
X_train_rfe1 = X_train_rfe.drop(['const'], axis=1)
```

```python

X_train_rfe1.columns
```

```python
# Calculating the VIFs for this model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe1
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

**observation** VIF is very high for the first model that we built leveraging top 20 ranked variables as per the RFE process
